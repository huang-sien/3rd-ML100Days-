```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.random(size=(10,10))-0.5

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
print(matrix)

heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap');
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.random(size=(nrow,ncol))-0.5

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
plot_data.head()
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(size=(nrow,ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
plot_data.head()
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
